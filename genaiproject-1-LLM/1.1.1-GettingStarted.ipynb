```python
import os
from dotenv import load_dotenv
load_dotenv()


""" by utilizing LANGCHAIN_TRACING_V2
Every LangChain call gets logged
You can see:
prompts
responses
latency
errors"""
## OpenAI API Key
openai_api_key = os.getenv("OPENAI_API_KEY")
if openai_api_key is not None:
    os.environ["OPENAI_API_KEY"] = openai_api_key

## Langsmith Tracking
langchain_api_key = os.getenv("LANGCHAIN_API_KEY")
if langchain_api_key is not None:
    os.environ["LANGCHAIN_API_KEY"] = langchain_api_key

os.environ["LANGCHAIN_TRACING_V2"] = "true"
langchain_project = os.getenv("LANGCHAIN_PROJECT")
if langchain_project is not None:
    os.environ["LANGCHAIN_PROJECT"] = langchain_project
```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model="gpt-4o")
print(llm)
```

```python
## Input and get response form LLM

result=llm.invoke("What is chatgpt40?")
print(result) 
print(type(result)) 
```

```python
print(type(result))
```

```python
### Chatprompt Template
from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_messages( 
    [
        ("system","You are an expert AI Engineer. Provide me answers based on the questions"),
        ("user","{input}")
    ]

)
prompt
```

```python
## chain    # asling a que to cchain by using llm
chain=prompt|llm

response=chain.invoke({"input":"Can you tell me about Langsmith?"})
print(response)
```

```python
print(type(response))
```

```python
## stroutput Parser

from langchain_core.output_parsers import StrOutputParser
output_parsers=StrOutputParser()
chain=prompt|llm|output_parsers

response=chain.invoke({"input":"Can you tell me about Langsmith?"})
print(response)
print(type(response))
```
